# Data Generator

```python
import random
import math

def rand_number():
    a = random.randint(0,9)*10**-1 + random.randint(0,9)*10**-2 + random.randint(0,9)*10**-3 + random.randint(0,9)*10**-4
    return a

def rand_sign():
    b = random.randint(0,1)
    if b == 0:
        c = -1
        return c
    if b == 1:
        c = 1
        return c
    
def sig_digit(a):
    sig_digits = 4
    rounded_number = round(a, sig_digits - int(math.floor(math.log10(abs(a))))-1)
    return rounded_number
```

```python
my_dict = {'p1':[],'p2':[],'p4':[],'p5':[],'p6':[],'p7':[],'T1':[],'T5':[],'T6':[],'T7':[]}
params = {'p1':0.01,'p2':0.4,'p4':5,'p5':4.8,'p6':4,'p7':3.8,'T1':42,'T5':130,'T6':400,'T7':380}
allowed_dev = 0.02

for x in my_dict:
    for y in range(172800):
        val = 0
        val = sig_digit(params[x]*(1+rand_sign()*rand_number()*allowed_dev))
#         data.append(val)
        my_dict[x].append(val)
```

```python
import pandas as pd

df = pd.DataFrame(my_dict)
df
```

```python
# Summary Statistics for each column

df.agg(
    {
        'p1': ['min','max','mean','median','std','skew'],
        'p2': ['min','max','mean','median','std','skew'],
        'p4': ['min','max','mean','median','std','skew'],
        'p5': ['min','max','mean','median','std','skew'],
        'p6': ['min','max','mean','median','std','skew'],
        'p7': ['min','max','mean','median','std','skew'],
        'T1': ['min','max','mean','median','std','skew'],
        'T5': ['min','max','mean','median','std','skew'],
        'T6': ['min','max','mean','median','std','skew'],
        'T7': ['min','max','mean','median','std','skew']        
    })
```
