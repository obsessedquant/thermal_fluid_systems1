# Real World Operational Considerations

In this section, the following topics are covered from real world operations:

* Routine start-up checks  
* Steady state operation  
* Special test, troubleshooting, diagnostics  

Failure modes:  

* Decreasing performance over time with sudden failure  
* Intermittent stalls with sudden failure  
* Measurement device failure

### Routine Start-up Checks

It's common for there to be a process and procedure to test all major components in a system prior to official start up.  

The example we're looking at here is for the high pressure pump.  

The pump is powered up to a lower pressure state and then held for a period of time. Then the pressure is bled off before going to a higher pressure state and held for a period of time.  

This is a typical preliminary step to check that the pump is operating correctly and that there are no unexpected leaks in the system.

### Failure Mode: Decreasing performance over time with sudden failure

This is a common failure mode observed in pressure containing systems. This failure mode is most often caused by mecahnical component failure or seal failure in pressure containing systems.  

![decreasing performance](Images/1_dec_perf_w_sudden_fail.png)

### Failure Mode: Intermittent Stall with Sudden Failure

This is a less common failure mode observed in pressure containing systems. This failure mode is most often caused by critical mecahnical component failure or a sliding seal failure in pressure containing systems.  

![intermittent stall failure](Images/1_intermittent_stall_w_sudden_fail.png)

### Failure Mode: Measurement Device Failure

This is a common failure mode observed in systems with measurement devices. This failure mode is caused by a failure of the measurement device itself. The failure can be either on the digital or the analog side.  

It's best to have two different measurement devices and methodologies for redundancies in monitoring of critical systems.  

![measurement devices](Images/1_measurement_devices_fail.png)
